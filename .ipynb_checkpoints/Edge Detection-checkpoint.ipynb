```python
import PIL.Image as Image

def save_imgs(imgs, folder_name):
    
    if not isinstance(folder_name, str):
        raise TypeError(
            f"String expected for folder_name, got '{type(folder_name).__name__}'"
        )
    
    if not isinstance(imgs, list | tuple):
        raise TypeError(
            f"list or tuple expected for folder_name, got '{type(imgs).__name__}'"
        )
    
    for i, img in enumerate(imgs):
        img.save(folder_name+ "_" + str(i) + '.png', format="PNG")

    return
```

```python
save_imgs(imgs, "trial")
```
